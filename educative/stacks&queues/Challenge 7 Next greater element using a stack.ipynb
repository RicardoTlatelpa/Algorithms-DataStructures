```python
def my_next_greater_element(lst):
    tmp_stack = MyStack()
    result_stack = MyStack()
    while(len(lst)):
        popped_value = lst.pop()
        while(tmp_stack.is_empty() == False):
            peeked = tmp_stack.peek()
            if(peeked > popped_value):
                result_stack.push(peeked)
                break
            else:
                tmp_stack.pop()
        if tmp_stack.is_empty():
            result_stack.push(-1)
        tmp_stack.push(popped_value)
    for i in range(result_stack.size()):
        lst.append(result_stack.pop())
    return lst
```

# Code anaylsis

So how does this algorithm work?

Well first, the first thing to recognize is that the input is a list not a stack, it's possible to pop but there are no helper stack methods

so as we pop the list

we append every popped value into a tmp_stack

and keep track of the next greater in the result_stack

The number one important method for this algorithm to work is the peek method

the peek method allows us to see the last popped value without popping from the stack

if the peeked value is greater than the current popped element from list than that means it's currently the greatest next number

if the peeked value is not greater than the currently popped value, then we pop from the tmp_list because we probably don't need this value anymore, it's the next greatest number

we keep popping until we find a greater number, if the tmp_list becomes empty, that means the current popped value is the next greater value for the remaining elements in the stack but also it is a -1 in the result stack

so we push -1 into the result list and then push the current popped value into the tmp list and the algorithm continues

This algorithm boils down to 3 situations, the peek value is either the next greatest number or the values potentially are, or the current popped value is the next greatest number
