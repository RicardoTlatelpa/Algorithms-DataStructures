unless we keep memory of the previous nodes in an array and pop off when we're triggered by the end of the list
because that means the node that is pointing to the end of the list going to be the head of the new reversed linked list

Going back to the same linked list 

1 -> 2 -> 3 -> 4 -> 5 -> 6 -> None

array_of_pointers = [1,2,3]

@ node 5 we can tell that the "next" node next node is the end of the list 6 -> None

take the current node, set its next node to None
and take the next node and set its next node to the current node





```python
def reverse(lst):
    previous = None
    current = lst.get_head()
    next = None
    
    while current:
        next = current.next_element
        current.next_element = previous
        previous = current 
        current = next
        
        lst.head_node = previous        
    return lst
```

# Code analysis 

So the way this solution works is very thought provoking but all of its operations are O(1)
within an O(N) loop, the most work we do in this solution is iterate through all the nodes in the linked list
and we have to because in order to have a reversed list, we have to touch all the elements to accomplish that.

But the way to think of every iteration is what surprised me. Every node is thought of the potential head of the final reversed linked list 

and its next_element pointer is dettached from each current node we traverse through. 

the node is isolated and then it's next element is immediately being attached to a previous node we have in memory
and then the node itself becomes the previous node in memory

As the list is traversed the line: lst.head_node = previous

makes each node feel like it can be the the final head node, it's not the final head node until we reach the final node

This sort of thinking of the linked list changes my perspective of only thinking of the linked list as a next_node.next_node.next_node chain, yes the only way we can see the elements in a linked list is by traversing linearly, but each node in a linked list can be isolated

and this pattern is seen when we delete nodes from a linked list, they are isolated as they are removed, the node that gets deleted has no next pointer and no node points to the deleted node, leaving the node out of the linked list

the linked list data structure is a chain of elements but it does't have to be thought of so rigidly as an array, a block of memory allocated elements, dynamically or not, the elements in an array stick in a rigid row together
